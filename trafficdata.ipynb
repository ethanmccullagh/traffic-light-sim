```python
import pandas
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
import random

data = pandas.read_csv('./clean_data.csv')

tStart = data['Start Time']


counts = []

for i in tStart: counts.append(0)

for i in ['North', 'East', 'South', 'West']:

    nS = data[i +' Straight']
    nR = data[i + ' Right']
    nL = data[i + ' Left']


    for i in range(len(tStart)):
        counts[i] += nS[i] + nR[i] + nL[i]
        

plt.plot(tStart, counts)
plt.xticks([i for i in range(0, 96, 12)])
plt.xlabel('Time of day')
plt.ylabel('Arrivals per 15 minutes')
plt.title('Daily arrivals')
plt.show()
```
